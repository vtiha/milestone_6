```python
import csv
from faker import Faker
import random

fake = Faker()
num_records = 55
department = ['HR','Finance','Engineering','R&D','Marketing','Sales','IT']

with open('database.csv', mode='w') as file:
    information = ['Full_Name', 'Hiring_Date', 'Department', 'Birthday', 'Phone_Number']
    writer = csv.DictWriter(file, fieldnames=information)
    writer.writeheader()
    for _ in range(num_records):
        writer.writerow({
            'Full_Name': fake.name(),
            'Hiring_Date': fake.date_this_decade(),
            'Department':random.choice(department),
            'Birthday': fake.date_of_birth(minimum_age=18, maximum_age=65),
            'Phone_Number': fake.numerify("380-9###-###-##")
        })

with open('database.csv', mode='r') as file:
    print(file.read())
```
